```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rado_f1 = pd.read_csv('./figure3/RADO_robust_f1.csv',index_col=0)
rado_auprc = pd.read_csv('./figure3/RADO_robust_auprc.csv',index_col=0)
rado_auroc = pd.read_csv('./figure3/RADO_robust_auroc.csv',index_col=0)
```

```python
logreg_f1 = pd.read_csv('./figure3/LogReg_robust_f1.csv',index_col=0)
logreg_auprc = pd.read_csv('./figure3/LogReg_robust_auprc.csv',index_col=0)
logreg_auroc = pd.read_csv('./figure3/LogReg_robust_auroc.csv',index_col=0)
```

```python
positive_ratio = logreg_f1.index / (logreg_f1.index+0.8)
```

```python
f1_data = pd.DataFrame({'imbalance ratio':positive_ratio,
                        'RADO':rado_f1.mean(axis=1), 
                        'Logistic Regression': logreg_f1.mean(axis=1)})
```

```python
auprc_data = pd.DataFrame({'imbalance ratio':positive_ratio,
                           'RADO':rado_auprc.mean(axis=1), 
                           'Logistic Regression': logreg_auprc.mean(axis=1)})
```

```python
auroc_data = pd.DataFrame({'imbalance ratio':positive_ratio,
                           'RADO':rado_auroc.mean(axis=1), 
                           'Logistic Regression': logreg_auroc.mean(axis=1)})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a color palette
rado_color = sns.color_palette("Set3")[0]
logreg_color = sns.color_palette("pastel")[1]

fig, axes = plt.subplots(1, 3, sharex=True, figsize=(12, 3))  # Adjusted figure size

# Plotting the data with specified colors
sns.lineplot(x='imbalance ratio', y='Logistic Regression', data=auroc_data, ax=axes[0], label='BCE loss', color=logreg_color)
sns.lineplot(x='imbalance ratio', y='RADO', data=auroc_data, ax=axes[0], label='AUPRC loss', color=rado_color)

sns.lineplot(x='imbalance ratio', y='Logistic Regression', data=auprc_data, ax=axes[1], label='BCE loss', color=logreg_color)  # No legend entry
sns.lineplot(x='imbalance ratio', y='RADO', data=auprc_data, ax=axes[1], label='AUPRC loss', color=rado_color)  # No legend entry

sns.lineplot(x='imbalance ratio', y='Logistic Regression', data=f1_data, ax=axes[2], label='BCE loss', color=logreg_color)  # No legend entry
sns.lineplot(x='imbalance ratio', y='RADO', data=f1_data, ax=axes[2], label='AUPRC loss', color=rado_color)  # No legend entry

# Hide top and right spines
for ax in axes:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.legend(loc='upper right')  # Positioning legend in upper right corner

# Set y-axis labels
axes[0].set_ylabel('AUROC')
axes[1].set_ylabel('AUPRC')
axes[2].set_ylabel('F1 Score')

# Set unified x-axis label and descending x-axis
for ax in axes:
    ax.set_xlim(0.5, 0.0)  # Adjusted to ensure all specified points are included
    ax.set_xlabel('Positive sample ratio')
    # ax.set_xticks([1.0, 0.75, 0.5, 0.25, 0.125])  # Uncommented to set x-axis tick positions

# Adjust layout
# plt.tight_layout()
fig.suptitle('Average performance on 18 datasets')
fig.savefig('figure3b.eps', format='eps',bbox_inches='tight')
# Show the plot
plt.show()
```
