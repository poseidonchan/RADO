```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./figure4.csv')
```

```python
data
```

```python
dataset1 = data.loc[data['Dataset']=='DOGMAseq1-ATAC-ch']
dataset2 = data.loc[data['Dataset']=='DOGMAseq-DIG-ATAC-ch']
```

```python
color_mapping = {
    'RADO': sns.color_palette('Set3')[3],
    'scDblFinder': sns.color_palette('Set3')[4],
    'AMULET':sns.color_palette('Set3')[5],
    'ArchR':sns.color_palette('Set3')[6],
}
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

sns.barplot(data=dataset1, x="Metric", y="Performance", hue='Method',
            width=0.8,
            hue_order=['RADO', 'scDblFinder', 'AMULET', 'ArchR'],
            palette=color_mapping, ax=axes[0])

sns.barplot(data=dataset2, x="Metric", y="Performance", hue='Method', 
            width=0.8,
            hue_order=['RADO', 'scDblFinder', 'AMULET', 'ArchR'],
            palette=color_mapping, ax=axes[1])

axes[0].set_xlabel('DOGMAseq1-ATAC-ch')
axes[1].set_xlabel('DOGMAseq-DIG-ATAC-ch')

for ax in axes:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_yticks([0,0.25,0.5,0.75,1])
    ax.legend(loc='upper right')  # Positioning legend in upper right corner
    
fig.savefig('figure4.eps', format='eps',bbox_inches='tight')
plt.show()
```
